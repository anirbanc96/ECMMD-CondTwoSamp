```python
import matplotlib.pyplot as plt
import seaborn as sns
import sbibm
```

```python
task = sbibm.get_task("two_moons")
reference = task.get_reference_posterior_samples(num_observation = 7)
```

```python
with sns.axes_style("white"):
    fig = plt.figure(figsize=(5,5))
    plt.title("True posterior", fontsize=15)
    plt.hexbin(reference[:,0], reference[:,1], gridsize=(150,150), cmap="Paired", rasterized=True)
    # plt.hist2d(reference[:,0], reference[:,1], bins=100, cmap="viridis", rasterized=True)
    # plt.axis("off")
    plt.yticks([])
    plt.xticks([])
    # plt.scatter(reference[:, 0], reference[:, 1], s = 0.5, alpha = 0.5)
    fig.savefig("reference_posterior7.pdf", transparent=True)
```
